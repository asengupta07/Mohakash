# Exploratory Data Analysis
In this notebook, we will perform some exploratory data analysis. The dataset used in this notebook is modified from the original dataset, which we created after feature extraction and dataframe formation from the logfile text document.

```python
import pandas as pd

pd.set_option('display.max_columns', None)
df = pd.read_csv('/content/drive/MyDrive/log_og v2.csv')
df
```

```python
def unique(df):
  for col in df.columns:
    print(f'{col}')
    print()
    print(df[col].value_counts())
    print('\n')
    print('_'*100)

unique(df)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.crscore.value_counts()
```

```python
df.crscore.fillna(0, inplace=True)
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df = df[~df.srcintf.isna()]
df.isna().sum()
```

```python
df.srccountry.fillna('Unknown', inplace=True)
df.isna().sum()
```

```python
df.duration.fillna(0, inplace=True)
df.sentbyte.fillna(0, inplace=True)
df.rcvdbyte.fillna(0, inplace=True)
df.isna().sum()
```

```python
df.appcat = df.appcat.apply(lambda x: 1 if x == 'unscanned' else 0)
```

```python
df.isna().sum()
```

```python
cp = df.drop(['srcport','dstport'], axis=1)
cp.isna().sum()
```

We have successfully removed and/or filled the missing values in the dataset. Thus we are left with a clean dataset, which can now be mapped and encoded in order to be used in our machine learning model.

# Encoding
In order to be able to use the dataset in our machine learning model, we need to encode the categorical features. For this, we will simply map the categorical features in the dataset to numerical values.

```python
unique(cp)
```

```python
cp.columns
```

```python
cp.type.unique()
```

```python
cp.type = cp.type.apply(lambda x: 1 if x == 'utm' else 0)
```

```python
cp.info()
```

```python
cp.subtype.unique()
```

```python
cols = cp.select_dtypes(include='object').columns.tolist()
d = {}
for col in cols:
  d[col] = cp[col].unique().tolist()
  li = cp[col].unique().tolist()
  cp[col] = cp[col].apply(lambda x: li.index(x))
d
```

```python
cp.info()
```

We now have successfully encoded the categorical features in the dataset. We can now save the dataset to a CSV file.

```python
cp.to_csv('df.csv', index=False)
```
